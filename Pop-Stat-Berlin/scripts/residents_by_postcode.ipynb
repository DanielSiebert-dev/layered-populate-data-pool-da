```python
import pandas as pd
import numpy as np
```

```python
# Load the CSV file
df = pd.read_csv('/Users/haritha/Downloads/Berlin_pop_stats - Residents by postcode,district and age group in 24.csv')
```

```python
df.head()
```

```python
# Convert the 'post_code' column to pandas nullable integer type 'Int64' to handle any missing values,
# then convert it to string type for consistent formatting
df['post_code'] = df['post_code'].astype('Int64').astype(str)
```

```python
df.info()
```

```python
cols_to_convert = ['total_residents', '0-6', '6-15', '15-18', '18-27',
                   '27-45', '45-55', '55-65', '65+', 'including_female']
```

```python
# Convert the columns in cols_to_convert to numeric types,
# then cast these columns to pandas nullable integer type 'Int64' to handle missing values properly.
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce').astype('Int64')
```

```python
df.info()
```

```python
df.head()
```

```python
# Drop the column 'including_female' from the DataFrame permanently
df.drop(columns='including_female', inplace=True)
```

```python
print(df['neighbourhood'].unique())
```

```python
# Standardize the 'neighbourhood' names by replacing abbreviated or inconsistent names
# with their full, lowercase standardized versions for uniformity across the dataset.
df['neighbourhood'] = df['neighbourhood'].replace({
    'Friedrichsh.-Kreuzb.': 'friedrichshainkreuzberg',
    'Charlbg.-Wilmersd.': 'charlottenburgwilmersdorf',
    'Tempelh.-Schöneb.': 'tempelhofschöneberg',
    'Steglitz-Zehlend.': 'steglitzzehlendorf',
    'Marzahn-Hellersd.': 'marzahnhellersdorf',
    'Treptow-Köpenick': 'treptowköpenick',
    'Neukölln': 'neukölln',
    'Pankow': 'pankow',
    'Lichtenberg': 'lichtenberg',
    'Reinickendorf': 'reinickendorf',
    'Mitte': 'mitte',
    'Spandau': 'spandau',
    'Berlin': 'berlin'
})
```

```python
print(df['neighbourhood'].unique())
```

```python
df.isnull().sum()
```

```python
# Remove rows where the 'post_code' value is missing (i.e., NaN or <NA>)
df = df[df['post_code'].notna()]
```

```python
df
```

```python
# Show all unique values in 'post_code' that are not numeric
df[~df['post_code'].str.fullmatch(r'\d{5}')]['post_code'].unique()
```

```python
# Replace '<NA>' string with actual NaN
df['post_code'] = df['post_code'].replace('<NA>', np.nan)
```

```python
# Now drop rows where post_code is missing
df = df[df['post_code'].notna()].reset_index(drop=True)
```
