```python
import pandas as pd
```

```python
# Load the CSV file
df = pd.read_csv('/Users/haritha/Downloads/Berlin_pop_stats - Berlin residents by district.csv')
```

```python
# Convert the 'last_updated_date' column to datetime format
df['last_updated_date'] = pd.to_datetime(df['last_updated_date'], errors='coerce')
```

```python
df.columns = df.columns.str.replace('\n', '', regex=True)
```

```python
df.columns
```

```python
# Clean column names by removing newlines and hyphens, then converting to lowercase
df.columns = df.columns.str.replace('\n', '', regex=True) \
                       .str.replace('-', '', regex=True) \
                       .str.lower()
```
