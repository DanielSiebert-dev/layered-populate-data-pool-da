```python
# import library to connect to api

import requests
```

```python
# initiate url for bus or tram stop data query

base_url = 'https://v6.bvg.transport.rest'

url_filter = '/locations?poi=false&address=false&query='
```

```python
# create function to retrieve information from api

def get_stop_data(stop_name):
    
    url = f"{base_url}{url_filter}{stop_name}"

    response = requests.get(url)

    if response.status_code == 200:
        return response.json()

    else:
        print(f"Retrieval failed. Status code: {response.status_code}")
```

```python
# call function to retrieve information about mehringerdamm

example = 'mehringerdamm'

mehringerdamm_data = get_stop_data(example)
```
