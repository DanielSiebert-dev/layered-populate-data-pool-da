### DATA PROCESSING

```python
# To define the important columns for the database
# These columns were chosen based on their importance for mapping and analysis

important_columns = [
    "Bad-ID",          # Unique ID for each pool/facility
    "Name",            # Facility name
    "Badtyp",          # Type (indoor, outdoor, open water, etc.)
    "Straße",          # Street address
    "Postleitzahl",             # Postal code
    "Ort",        # Local neighborhood / quarter
    "Bezirk",          # Borough / district
    "Breitengrad",        # Latitude for map placement
    "Längengrad",       # Longitude for map placement
    "eu_badegewaesser",# EU-designated bathing site flag
    "Barrierefreiheit",# Accessibility info
    "Wasserqualitaet", # Water quality rating
    "Öffnungsstunden pro Jahr",  # Opening hours (if present in dataset)
    "Baujahr",          # Year built (for historical context)
    "Ganzjährig geöffnet",
    "Name des Eigentümers",
    "Ermäßigung Kind",  # Discount for children
    "Ermäßigung Familie",  # family discount
    "Ermäßigung Behinderte",  # discount for behinder
    
]
```

```python
important_columns
```

```python
# To Keep only these columns from the  dataset for further assessment
require_columns = [col for col in important_columns if col in df.columns]
filtered_df = df[require_columns]

require_columns
```

```python
# I save these columns for further data modelling 
filtered_df.to_csv("pools_berlin_citymap.csv", index=False)
```

### Berlin pools data further exploration

```python
df_new = pd.read_csv("pools_berlin_citymap.csv")
```

```python
df_new.head()
```

```python
df_new.columns = df_new.columns.str.lower().str.replace(' ', '_')
```

```python
df_new.head()
```

```python
# translating german column names to english
rename_map = {
    'bad-id': 'pool_id',
    'name': 'name',
    'badtyp': 'pool_type',
    'straße': 'street',
    'postleitzahl': 'postal_code',
    'ort': 'city',
    'breitengrad': 'latitude',
    'längengrad': 'longitude',
    'eu_bathing_water': 'eu_bathing_water',
    'öffnungsstunden_pro_jahr': 'annual_opening_hours',
    'baujahr': 'year_built',
    'ganzjährig_geöffnet': 'open_all_year',
    'name_des_eigentümers': 'owner_name',
    'ermäßigung_kind': 'discount_children',
    'ermäßigung_familie': 'discount_family',
    'ermäßigung_behinderte': 'discount_disabled'
}

df_new.rename(columns=rename_map, inplace=True)

print(df_new.columns.tolist())


```

```python
df_new.head()
```

```python
df_new.info()
```

```python

df_new.duplicated(subset='pool_id').sum()  # checking for duplicate
```

```python
df_new.info()
```

```python
df_new['postal_code'] = df_new['postal_code'].astype(str)

```

```python
df_new['city'] = df_new['city'].str.title()
```

```python
df_new.info()
```

```python

# Convert yes/no or German terms to Boolean
yes_values = ['Ja', 'Yes', 'yes', 'ja', True]
df_new['open_all_year'] = df_new['open_all_year'].apply(lambda x: True if x in yes_values else False)
```

```python
# To keep only valid geographic coordinates

df_new = df_new[(df_new['latitude'].between(-90, 90)) & (df_new['longitude'].between(-180, 180))]
```

```python
# Selecting these columns ensure the dataset is both geospatially accurate and informative for analysis, 
# making it suitable for interactive maps, location clustering, and accessibility studies.

mapping_cols = [
    'pool_id', 'name', 'pool_type', 'street', 'postal_code', 'city',
    'latitude', 'longitude', 'open_all_year'
]
df_new = df_new[mapping_cols]
```

```python
df_new['street'] = df_new['street'].fillna("Unknown")
```
